# Linear Regresion scikitlearn

The objective of this project is to predict wine quality taking into account its components. 

###### Import pandas and read data base

```python
import pandas as pd
wine = pd.read_csv("C:\\Users\\DEAM\\Downloads\\winequality-white.csv", sep=';')
wine
```

Division of the data as all data comes from de same dataframe

```python
components = list(wine.columns)
components.remove('quality')

x = wine[components]
y = wine['quality']
```

Import scikitlearn libraries

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y)
```

creating the model

```python
model = LinearRegression()
model.fit(x_train, y_train)

training_pred = model.predict(x_train)
test_pred = model.predict(x_test)

print('Training R2 is: ', model.score(x_train, y_train))
print('Test R2 is: ', model.score(x_test, y_test))
```
